###**Importing the required modules**

```python
# import relevant libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import csv
import re
import pickle
```

###**Loading in the csv**

```python
df = pd.read_csv('/content/drive/MyDrive/Webshop_classification/data/clean_pred_dataset.csv')
```

```python
# print the first rows of the DataFrame
df.head()
```

### **Loading in the pickle**

```python
# load model from drive
filename = '/content/drive/MyDrive/Webshop_classification/data/best_model.p'
best_model = pickle.load(open(filename, 'rb'))
```

### **Make predictions**

```python
# define the variables used for predictions
names = np.array(df['Website'])
X_pred = np.array(df['Clean_BoW'])
y_pred = best_model.predict(X_pred)
```

```python
# check the number of labels in the array 
print(f'The model predicted {len(y_pred)} labels')
```

```python
# create final dataframe containing the websites and the corresponding labels
df_pred = pd.DataFrame(list(zip(names, y_pred)), columns=['domain', 'is_shop'])
df_pred.head()
```

### **Exporting to csv**

```python
# export to csv-file
df_pred.to_csv('/content/drive/MyDrive/Webshop_classification/data/final_dataset.csv', index=False)
```
